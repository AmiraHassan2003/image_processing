```python
import tkinter as tk
from tkinter import filedialog, messagebox
from PIL import Image, ImageTk

class ImageNumberApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Image and Number Input")
        
        # Create a button to upload an image
        self.upload_button = tk.Button(root, text="Upload Image", command=self.upload_image)
        self.upload_button.pack()

        # Label to display the uploaded image
        self.image_label = tk.Label(root)
        self.image_label.pack()

        # Entry to input a number
        self.number_entry = tk.Entry(root)
        self.number_entry.pack()

        # Button to submit the number
        self.submit_button = tk.Button(root, text="Submit Number", command=self.submit_number)
        self.submit_button.pack()

    def upload_image(self):
        file_path = filedialog.askopenfilename()
        if file_path:
            # Load and display the image
            self.image = Image.open(file_path)
            self.image = self.image.resize((250, 250), Image.ANTIALIAS)  # Resize for display
            self.photo = ImageTk.PhotoImage(self.image)
            self.image_label.config(image=self.photo)
            self.image_label.image = self.photo  # Keep a reference to avoid garbage collection

    def submit_number(self):
        number = self.number_entry.get()
        if number.isdigit():
            messagebox.showinfo("Submitted", f"You have submitted the number: {number}")
        else:
            messagebox.showerror("Error", "Please enter a valid number.")

name = "main"
if name == "main":
    root = tk.Tk()
    app = ImageNumberApp(root)
    root.mainloop()
```

 # Grayscale

```python
def grayscale_image(imagePath):
    input_image = Image.open(imagePath)
    grayscale_image = input_image.convert('L')
    # print(input_image)
    # grayscale_image.save(filepath)
    return grayscale_image
# grayscale_image(filepath)
```

 # Threshold 

```python
def calculate_average_threshold(imagePath):
    grayscale_image_pixel = np.array(grayscale_image(imagePath))
    flattened_matrix = grayscale_image_pixel.flatten()
    threshold_average = sum(flattened_matrix) // len(flattened_matrix)
    return threshold_average
```

# Simple Halftoning  

```python
def simple_halftone(imagePath):
    threshold = calculate_average_threshold(imagePath)
    grayscale_image_pixel = np.array(grayscale_image(imagePath))
    # simple_halftone_image = grayscale_image_pixel.copy()
    # print(grayscale_image_pixel)
    # print()
    for row in range(len(grayscale_image_pixel)):
        for col in range(len(grayscale_image_pixel[0])):
            if grayscale_image_pixel[row,col] < threshold:
                grayscale_image_pixel[row,col] = 0
            else:
                grayscale_image_pixel[row,col] = 1
            # simple_halftone_image[row][col] = grayscale_image_pixel[row][col]
    # print(grayscale_image_pixel)
    return grayscale_image_pixel
    
```

 # Advanced Halftoning 

```python
def advanced_halftone(imagePath):
    threshold = calculate_average_threshold(imagePath)
    grayscale_image_pixel = np.array(grayscale_image(imagePath))
    # print(grayscale_image_pixel)
    # print()
    
    for row in range(len(grayscale_image_pixel)):
        for col in range(len(grayscale_image_pixel[0])):
            old_pixel = grayscale_image_pixel[row][col]
            # print(f"old_pixel: {old_pixel}")
            
            if grayscale_image_pixel[row][col] < threshold:
                grayscale_image_pixel[row][col] = 0
            else:
                grayscale_image_pixel[row][col] = 255 
                
            new_pixel = grayscale_image_pixel[row][col]
            # print(f"new_pixel: {new_pixel}")
            
            error = old_pixel - new_pixel
            # print(f"error: {error}")
            
            if (col+1) < len(grayscale_image_pixel[0]):
                grayscale_image_pixel[row][col+1] += (7/16) * error
            
            if ((row+1) > len(grayscale_image_pixel)) & ((col+1) > len(grayscale_image_pixel[0])):
                grayscale_image_pixel[row+1][col+1] += (1/16) * error

            if (row+1) > len(grayscale_image_pixel):
                grayscale_image_pixel[row+1][col] += (5/16) * error
            
            if ((row+1) > len(grayscale_image_pixel)) & ((col-1) < len(grayscale_image_pixel[0])):
                grayscale_image_pixel[row-1][col-1] += (3/16) * error 
            
            
            
            # simple_halftone_image[row][col] = grayscale_image_pixel[row][col]
    # print(grayscale_image_pixel)
    return grayscale_image_pixel
```

 # Histogram 

```python
def histogram(imagePath):
    grayscale_image_pixel = np.array(grayscale_image(imagePath))
    repeat_color_dict = {}
    height, width = grayscale_image_pixel.shape
    area_image = height * width

    for row in range(len(grayscale_image_pixel)):
        for col in range(len(grayscale_image_pixel[0])):
            pixel_value = grayscale_image_pixel[row, col]
            if pixel_value in repeat_color_dict:
                repeat_color_dict[pixel_value] += 1
            else:
                repeat_color_dict[pixel_value] = 1
    
    pixel_values = list(repeat_color_dict.keys())
    frequencies = list(repeat_color_dict.values())
    # print(f"frequencies: {frequencies}")
    # return max(pixel_values)

    # Plotting the histogram
    plt.bar(pixel_values, frequencies, width=1, color='black')
    plt.title('Histogram of Grayscale Image')
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.xlim(0, 255)
    plt.show()
```

 # Histogram Equalization

```python
def histogram_equalization(imagePath):
    grayscale_image_pixel = np.array(grayscale_image(imagePath))
    repeat_color_dict = {}
    modified_frequencies = []
    new_frequency = 0
    height, width = grayscale_image_pixel.shape
    image_area = height * width
    
    for row in range(len(grayscale_image_pixel)):
        for col in range(len(grayscale_image_pixel[0])):
            pixel_value = grayscale_image_pixel[row, col]
            if pixel_value in repeat_color_dict:
                repeat_color_dict[pixel_value] += 1
            else:
                repeat_color_dict[pixel_value] = 1
    
    pixel_values = list(repeat_color_dict.keys())
    frequencies = list(repeat_color_dict.values())
    max_level = max(pixel_values)
    # print("max_key: ",max_level)
    
    for key, value in repeat_color_dict.items():
        # print(f"key: {key}\n value: {value}")
        new_frequency += max_level * (1 / image_area) * value
        modified_frequencies.append(new_frequency)
    # print(f"modified_frequencies: {modified_frequencies}")
    
    # Plotting the histogram
    plt.bar(pixel_values, modified_frequencies, width=1, color='black')
    plt.title('Histogram Equalization of Grayscale Image')
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.xlim(0, 255)
    plt.show()
    
```
